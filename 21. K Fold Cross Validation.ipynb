```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import load_digits
import pandas as pd

digits = load_digits()
```

### Need for K Fold Validation

- During splitting data into train and tests train_test_split splits randomly.
- Accuracy depends on data
- Using K Fold validation dataset can be divided as many times as required and accuracy is predicted.

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
```

```python
for train_index, test_index in kf.split([1,2,3,4,5,6,7,8,9,0]):
    print(train_index, test_index)
```

```python
def getScore(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
logisticRegression = []
svm = []
randomForestClassifier = []

for train_index, test_index in kf.split(digits.data):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
    digits.target[train_index], digits.target[test_index]
    
    logisticRegression.append(getScore(LogisticRegression(max_iter=3000), X_train, X_test, y_train, y_test))
    svm.append(getScore(SVC(), X_train, X_test, y_train, y_test))
    randomForestClassifier.append(getScore(RandomForestClassifier(), X_train, X_test, y_train, y_test))
```

```python
print(logisticRegression,svm,randomForestClassifier)
```

## StratifiedKFold

- There may be cases where classes aren't distributed in the train test split dataframes so StratifiedKFold ensures that classes are distributed in both test and train datasets.

```python
from sklearn.model_selection import StratifiedKFold
fold = StratifiedKFold(n_splits=3)
```

```python
logisticRegression = []
svm = []
randomForestClassifier = []

for train_index, test_index in fold.split(digits.data, digits.target):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
    digits.target[train_index], digits.target[test_index]
    
    logisticRegression.append(getScore(LogisticRegression(max_iter=30000), X_train, X_test, y_train, y_test))
    svm.append(getScore(SVC(), X_train, X_test, y_train, y_test))
    randomForestClassifier.append(getScore(RandomForestClassifier(), X_train, X_test, y_train, y_test))
```

```python
print(logisticRegression,svm,randomForestClassifier)
```

- All the above process is bundled into cross_val_score

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(max_iter=30000), digits.data, digits.target)
```

```python
cross_val_score(SVC(), digits.data, digits.target)
```

```python
cross_val_score(RandomForestClassifier(), digits.data, digits.target)
```

# Linear Regression

<img src="./assets/01.png" />

- We square the distance between points and the line, sum it up and try to minize the sum by considering different lines. The line which has minimum sum is considered as best fit line. 

<img src="./assets/02.png" />

- Here y is dependent variable (i.e, price) and x is independent variable (i.e, area)
- Modal basically tries to calculate m (slope also called as coeff) and intercept (b)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Load data frame
df = pd.read_csv('./prices.csv')
```

```python
# Plot data points

# To draw static image in notebook
%matplotlib inline 
plt.xlabel('Area(sqft)')
plt.ylabel('Prices($)')
plt.scatter(df.area, df.price, color='red', marker='x')
```

```python
# model
linearRegression = linear_model.LinearRegression()
linearRegression.fit(df[['area']], df.price)
```

```python
linearRegression.predict(np.array([[3300]]))
```

```python
print({
    "co-efficient": linearRegression.coef_[0],
    "intercept": linearRegression.intercept_
})
```

```python
m = linearRegression.coef_[0]
b = linearRegression.intercept_

y = m*(3300)+b;
y # which exactly matches what is predicted by Linear Regression model
```

```python
plt.scatter(df['area'], df['price'], color='red', marker='x')

areas = np.array([3300,2700,2300,4500])
plt.plot(areas, linearRegression.predict(areas.reshape(-1,1)))
```

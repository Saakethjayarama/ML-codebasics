# Pickle

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
# Load data frame
df = pd.read_csv('./prices.csv')
```

```python
linearRegression = linear_model.LinearRegression()
linearRegression.fit(df[['area']], df.price)
```

```python
import pickle

with open('model.pkl', 'wb') as f:
    pickle.dump(linearRegression, f)
```

```python
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)
    x = model.predict(np.array([[5000]]))
    print(x)
```

# Joblib

```python
from joblib import load, dump
```

```python
dump(linearRegression, 'model_job')
```

```python
model = load('model_job')
model.predict(np.array([[5000]]))
```

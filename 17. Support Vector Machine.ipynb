# Support Vector Machines

- Maximize distance between points and the line

- 3D - Plane
- nD - Hyper plane
- Gamma and Regularization

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='red', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='green', marker='x')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='blue', marker='*')
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='red', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='green', marker='x')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='blue', marker='*')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target', 'flower_name'], axis='columns'), df['target'])
```

```python
from sklearn.svm import SVC
model = SVC(C=2) 
# regularization C, gamma and kernel
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

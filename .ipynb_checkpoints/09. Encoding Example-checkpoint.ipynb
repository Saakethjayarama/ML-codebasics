```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
# Plot

%matplotlib inline

plt.xlabel('mileage')
plt.ylabel('sell price')
plt.scatter(df['Mileage'], df['Sell Price($)'])

# From graph we can deduce that sell price and mileage has linear relationship
```

```python
plt.xlabel('Age')
plt.ylabel('Sell price')
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])

# from the below graph we can deduce that age and sell price has linear relationship
```

- Now that we know the dependent and independent variables have linear relationship we can use linear models

```python
le = LabelEncoder()
df['Car Model'] = le.fit_transform(df['Car Model'])
df
```

```python
y = df['Sell Price($)']
y
```

```python
ohe = OneHotEncoder()
car_models = pd.DataFrame(ohe.fit_transform(df[['Car Model']]).toarray())
car_models = car_models.iloc[:, 1:]
car_models
```

```python
X = df.drop(['Sell Price($)', 'Car Model'], axis='columns')
X
```

```python
X = pd.concat([car_models, X], axis='columns')
X
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# 4yr old Mercedez benz with mileage 45000
model.predict([[0,1, 45000, 4]])
```

```python
# 7yr old BMW with mileage 86000
model.predict([[1,0, 86000, 7]])
```

```python
model.score(X, y)
```

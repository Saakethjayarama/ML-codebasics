```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('perCapita.csv')
```

```python
# plot
%matplotlib inline

plt.xlabel('year')
plt.ylabel('Per capita income ($)')
plt.scatter(df[['year']], df[['per capita income (US$)']])
```

```python
#Train
linearRegression = LinearRegression()
linearRegression.fit(df[['year']], df['per capita income (US$)'])
```

```python
# Inference
linearRegression.predict([[2021]])
```

```python
# Visualizing

years = np.array([1980,1990,2000,2010,2020,2030])


plt.xlabel('year')
plt.ylabel('Per capita income ($)')
plt.scatter(df[['year']], df[['per capita income (US$)']], color='red', marker='*')

plt.plot(years.reshape(-1,1), linearRegression.predict(years.reshape(-1,1)))
plt.scatter([[2021]], linearRegression.predict([[2021]]), color='black')
```

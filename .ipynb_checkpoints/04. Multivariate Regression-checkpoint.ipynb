- price = m1*area+m2*bedrooms+m3*age+b
- where area, bedrooms and age are independent variables or features. m1, m2 and m3 are coefficients and b is an intercept.
- Model building for this dataset involves to step
    1. Data preprocessing : Handling NA values
    2. Linear regression using multiple variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
# Data pre processing
import math
median = math.floor(df.bedrooms.median())

df.bedrooms = df.bedrooms.fillna(median)
df
```

```python
# Train model

linearRegression = LinearRegression()
data = df[['area', 'bedrooms', 'age']]
data
```

```python
linearRegression.fit(data, df['price'])
print({
    "coefficients": linearRegression.coef_,
    "interepts": linearRegression.intercept_
})
```

```python
linearRegression.predict([[3000,3,40], [2500,4,5]])
```
